```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
def generate_signal(n:int):
    x = np.random.rand(n)
    w = 2*np.pi/n
    for m_i in range(x.shape[0]):
        x[m_i] = np.sin(w*12*m_i) + np.cos(w*19*m_i)
    return x
```

```python
if __name__ == "__main__":
    
    data = generate_signal(143)
    fft_result = np.fft.fft(data)
```

```python
plt.plot(data)
```

```python
plt.plot(np.abs(fft_result[0:70]))
```

```python
h = scipy.signal.hilbert(data)
```

```python
for i, v in enumerate(h):
    print(i,v )
```

```python
plt.plot(np.real(h))
plt.plot(np.imag(h))
```
